```python
import json
import os
import pandas as pd
```

```python
with open("../../data/prediction_val_sub01/val_predict_sub01_finetuned_fixed_id.json", "r") as f_q:
    answers_val = json.load(f_q)
```

```python
val_dataset = pd.read_csv("./vqa_val_sub01.tsv", sep="\t", 
                          names = ["question_id", "image-id", "question", "answer", "labels", "image"])
val_dataset.drop(["labels", "image"], inplace = True, axis=1)
val_dataset["answer"] = val_dataset.apply(
    lambda x: x["answer"].split('|!+')[1] , axis=1)
```

```python
val_dataset
```

```python
answers_df = pd.DataFrame(answers_val)
joined = answers_df.join(val_dataset.set_index("question_id"),how = "inner", on = "question_id",lsuffix="-generated", rsuffix="-gt")
cols = ["question_id", "image-id", "question", "answer-gt","answer-generated"]
joined = joined[cols]
```

```python
joined.to_csv("../../data/prediction_val_sub01/preditctions_gt_q.csv")
```
